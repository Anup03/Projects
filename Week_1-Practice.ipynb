```python
import pandas as pd

step_data = [1234, 4563, 6532, 2134, 5342, 1236]
step_counts = pd.Series(step_data, name='steps')
print step_counts
```

```python
step_counts.index = pd.date_range('20180506', periods=6)
```

```python
print step_counts
```

```python
print step_counts['20180506']
print step_counts[4]
print step_counts['2018-05']
```

```python
# View the data type
step_counts.dtypes
```

```python
# Converting data type
import numpy as np
step_counts = step_counts.astype(np.float)
print step_counts.dtypes
```

```python
# Create invalid data
step_counts[1:3] = np.NaN
```

```python
step_counts
```

```python
# Fill nan with zeros
step_counts = step_counts.fillna(0.)
```

```python
step_counts
```

```python
# In single line
step_counts.fillna(0., inplace=True)
```

```python
step_counts
```

```python
filepath = 'data/Iris_Data.csv'
data = pd.read_csv(filepath)
print data.iloc[:5]
```

```python
import os
print os.getcwd()
```

```python
print os.listdir(os.getcwd())
```

```python
# Creaate a new column that is product of both dimensions
data['sepal_area'] = data.sepal_length * data.sepal_width

# print a few rows & columns
print(data.iloc[:5, -3:])
```

```python
# lambda function applies what follows it to each row of data
data['abbrev'] = (data
                 .species
                 .apply(lambda x:
                       x.replace('Iris-','')))
```

```python
print data.iloc[:5, -3:]
```

```python
# Concatenate first two and last two rows
small_data = pd.concat([data.iloc[:2],
                      data.iloc[-2:]])
```

```python
print small_data.iloc[:, -3:]
```

```python
# Using the groupby method calculates aggregated DataFrame statistics
group_sizes = (data.
              groupby('species')
              .size())
```

```python
print group_sizes
```

```python
print data.groupby('species').size()
```

```python
# Mean calculated on a DataFrame
print data.mean()
```

```python
# Median calculated on a Series
print(data.petal_length.median())
```

```python
# mode calculated on a Series
print(data.petal_length.mode())
```

```python
# Standard Deviation, variance, SEM and quantiles can also be calculate
print(data.petal_length.std(),
      data.petal_length.var(),
      data.petal_length.sem())
```

```python
print data.quantile(0)
```

```python
# Multiple calculations can be presented in DataFrame
print data.describe()
```

```python
# DataFrames can be randomly sampled from
# Sample 5 rows without replacement
sample = (data
         .sample(n=5,
                replace=False,
                random_state=0))
```

```python
print sample.iloc[:, -3:]
```

```python
# Scatter plots can be created from Pandas Series
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(data.sepal_length,
        data.sepal_width,
        ls='', marker='o')
```

```python
# Multiple layers of data can also be added
plt.plot(data.sepal_length,
        data.sepal_width,
        ls='', marker='o',
        label='sepal')
plt.plot(data.sepal_length,
        data.sepal_width,
        ls='', marker='o',
        label='petal')
```

```python
# Histograms can be crated from Pandas
plt.hist(data.sepal_length, bins=25)
```

```python
fig, ax = plt.subplots()
ax.barh(np.arange(10),
       data.sepal_width.iloc[:10])
```

```python
fig, ax = plt.subplots()
ax.barh(np.arange(10),
       data.sepal_width.iloc[:10])
# Set position of ticks and tick labels
ax.set_yticks(np.arange(0.4, 10.4, 1.0))
ax.set_yticklabels(np.arange(1,11))
ax.set(xlabel='xlabel', ylabel='ylabel', title='Title')
```

```python
# statistical cal can be included with pandas 
(data.groupby('species')
.mean()
.plot(color=['red','blue','black','green'],
     fontsize=10.0, figsize=(4,4)))
```

```python
# Joint distribution and scatter plots can be created
%seaborn inline
import seaborn as sns
sns.jointplot(x='sepal_length',
             y='sepal_width',
             data=data, size=4)
```

```python
# Correlation plots of all variable pair can also be made with seaborn
sns.pairplot(data, hue='species', size=3)
```
